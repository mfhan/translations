```python
import torch
import pandas as pd
from transformers import pipeline
import re
```

```python
edf2 = pd.read_csv("proust/vinteuil2_english.csv", index_col = 0)
edf2.head()
```

```python
enresults = []
# for sentences in df:
for index, row in edf2.iterrows():
    dict = {}
    sentiment_pipeline = pipeline("sentiment-analysis", model="cardiffnlp/twitter-xlm-roberta-base-sentiment")
    data = index 
    rating = sentiment_pipeline(data)
    # print(index)
    # print(rating)
    #df['score'] = rating
    # print(df['score'])
    dict['text'] = index
    dict['label'] = rating[0]['label']
    dict['score'] = rating[0]['score']
    enresults.append(dict)
print(enresults[0]) 
```

```python
engdf = pd.DataFrame(enresults)
engdf
```

```python
engdf['label'].value_counts()
```

```python
fdf = pd.read_csv("proust/vinteuil2_french.csv", index_col = 0)
fdf.head()
```

```python
fresults = []
for index, row in fdf.iterrows():
    dict = {}
    sentiment_pipeline = pipeline("sentiment-analysis", model="cardiffnlp/twitter-xlm-roberta-base-sentiment")
    data = index 
    rating = sentiment_pipeline(data)
    #df['score'] = rating
    # print(df['score'])
    dict['text'] = index
    dict['label'] = rating[0]['label']
    dict['score'] = rating[0]['score']
    fresults.append(dict)
```

```python
print(fresults[0])
```

```python
frenchdf = pd.DataFrame(fresults)
frenchdf
```

```python
frenchdf['label'].value_counts()
```

```python
frenchdf.groupby(['label'])['score'].mean()
```

```python
engdf.groupby(['label'])['score'].mean()
```

```python
for row in fresults: 
    if row['label'] == 'negative':
        if row['score'] > 0.5:
            row['categ_score'] = -2
        else: 
            row['categ_score'] = -1
    elif row['label'] == 'positive':
        if row['score'] > 0.5:
            row['categ_score'] = 2
        else:
            row['categ_score'] = 1
    else:
        row['categ_score'] = 0
print(fresults[0])            
```

```python
fr_categ = pd.DataFrame(fresults)
fr_categ.head()
```

```python
fr_categ.to_csv("proust_fr_categ.csv")
```
